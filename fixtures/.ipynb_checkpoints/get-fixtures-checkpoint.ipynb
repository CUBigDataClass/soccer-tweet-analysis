```python
import pandas as pd
from datetime import datetime
import json
```

```python
df0 = pd.read_csv("E0.csv")
df1 = pd.read_csv("E1.csv")
df = df0.append(df1)
```

```python
df.head()
```

```python
df['Date'] = df['Date'].apply(lambda x : datetime.strptime(x, '%d/%m/%y'))
```

```python
df.head()
```

```python
mufc = df[df['HomeTeam'].str.contains("Man United")].append(df[df['AwayTeam'].str.contains("Man United")])
liverpool = df[df['HomeTeam'].str.contains("Liverpool")].append(df[df['AwayTeam'].str.contains("Liverpool")])
mcfc = df[df['HomeTeam'].str.contains("Man City")].append(df[df['AwayTeam'].str.contains("Man City")])
arsenal = df[df['HomeTeam'].str.contains("Arsenal")].append(df[df['AwayTeam'].str.contains("Arsenal")])
spurs = df[df['HomeTeam'].str.contains("Tottenham")].append(df[df['AwayTeam'].str.contains("Tottenham")])
chelsea = df[df['HomeTeam'].str.contains("Chelsea")].append(df[df['AwayTeam'].str.contains("Chelsea")])
```

```python
mufc_fixtures = list()
from collections import OrderedDict

for idx, val in mufc.iterrows():
    d = OrderedDict()
    d['date'] = val[1]
    d['club1'] = val[2]
    d['club1score'] = val[4]
    d['club2'] = val[3]
    d['club2score'] = val[5]
    mufc_fixtures.append(d)
    
mufc_fixtures = sorted(mufc_fixtures, key=lambda x : x['date'])
```

```python
for i in mufc_fixtures:
    i['date'] = i['date'].strftime('%m/%d/%Y')

with open('mufc_fixtures','w') as f:
    json.dump(mufc_fixtures, f)
```
